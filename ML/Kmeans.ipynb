```python
import random
```

```python
def random_centers(X,k):
    center=[]
    for i in range(k):
        rand= random.randint(0,len(X))
        center.append(X[rand])
    return center
```

```python
def dist(a,b):
    d = ((b[0]-a[0])**2+(b[1]-a[1])**2)**.5
    return d
```

```python
def mean_center(x,l):
    new_cs=[]
    for i in range(l):
        new_c=[]
        np = 0
        totalp=[]
        for j in x:
            if j[-1]==i:
                new_c.append(j)
        sumx=0     
        sumy=0
        for k in new_c:
            sumx += k[0]
            sumy += k[1]
            mc = [sumx/len(new_c),sumy/len(new_c)]
        new_cs.append(mc)
    return new_cs
```

```python
def cluster(x,c,fc = False):
    #
    for i in x:
        dis =99999
        ce =0
        arr = []
        for j in range(len(c)):
            d = dist(i,c[j])
            if d<dis:
                dis = d
                ce = j
        if fc:
            i.append(ce)
        else:
            i[-1]= ce
            
    return x,len(c)
        
```

```python
def train(X,k,n):
    c = random_centers(X,k)
    clus,l = cluster(X,c,fc=True)
    for i in range(n):
        centers = mean_center(clus,l)
        clus,l = cluster(X,centers,fc=False)
    return centers
    
```

```python
centers = train(X,3,8)
```

```python
from sklearn.cluster import KMeans
```

```python
inertias=[]
for i in range(1,11):
    k_means= KMeans(n_clusters=i)
    k_means.fit(x)
    inertias.append(k_means.inertia_)
```

```python
plt.plot(inertias)
```
