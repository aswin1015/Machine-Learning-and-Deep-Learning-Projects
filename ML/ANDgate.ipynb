```python
import tensorflow as tf
from keras import models, layers, Sequential
import numpy as np
```

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([[0],[0],[0],[1]])
```

```python
model= models.Sequential()
```

```python
model.add(layers.Dense(16,activation='relu',use_bias=False))
model.add(layers.Dense(1,activation='sigmoid',use_bias=False))
```

```python
model.compile(
    optimizer='adam',
    loss='mse')
```

```python
model.fit(x,y,epochs=500)
```

```python
l=model.predict(np.array([[1,1]]))
```

```python
print(1 if l>0.5 else 0)
```
