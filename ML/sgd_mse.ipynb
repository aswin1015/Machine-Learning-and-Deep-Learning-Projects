```python
data=[15,101,18,7,13,16,11,21,5,25,10,9]
import statistics as st
print(st.stdev(data))
```

```python
import pandas as pd
import numpy as np


def out1(data):
    tresh=3
    mean= np.mean(data)
    std= np.std(data)
    outliers=[]
    for i in data:
        score=(i-mean)/std
        if np.abs(score)> tresh:
            outliers.append(i)
    print(outliers)
        
    
```

```python
out1(data)
```

```python
tenth_percentile=np.percentile(data,10)
ninty_percentile=np.percentile(data,90)
b= np.where(data<tenth_percentile,tenth_percentile,data)
b= np.where(data>ninty_percentile,ninty_percentile,data)
print(b)
```

```python
q1= np.percentile(data,25)
q3=np.percentile(data,75)

iqr=q3-q1
```

```python
lrb= q1-(1.5*iqr)
upr = q3+(1.5*iqr)
out1=[]
```

```python
for i in data:
    if i<lrb or i >upr:
        out1.append(i)
```

```python
print(out1)
```
